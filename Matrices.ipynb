```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
data1
```

# Ejercicio 1: 

Demuestra que la matriz de correlación es una matriz simétrica por los siguientes métodos

*Usando Trnaspuesta

*Usando las matrices triangulares

```python
# usando transpuesta
data_cor = data1.corr()
data_cor_t = data_cor.T
#comparar
data_cor == data_cor_t

# funcion de numpy
np.array_equal(data_cor, data_cor_t)
```

```python
data_cor
```

```python
#2 
tri_sup = np.triu(data_cor, 0 )
tri_inf = np.tril(data_cor, -1)
#primer intento
tri_inf + tri_sup == data_cor
# segundo intento
np.array_equal(data_cor, tri_inf + tri_sup )
```

```python
# tercera forma
tri_sup = np.triu(data_cor, 0 )
tri_inf = np.tril(data_cor, 0)
tri_sup == tri_inf.T
```

# Ejercicio 2
Obten la matriz inversa de la matriz de correlación

```python
data_cor_inv = np.linalg.inv(data_cor)

# metodo : adjunta / determinante 
np.linalg.det(data_cor) 
```
